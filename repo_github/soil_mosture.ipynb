```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

df = pd.read_csv(r'C:\Users\user\.kaggle\soil_moisture_1_Raw.csv')
print("Data mentah berhasil dibaca!")
print(f"Shape: {df.shape}")
print(df.head())

df_clean = df.copy()

moisture_cols = ['moisture0', 'moisture1', 'moisture2', 'moisture3', 'moisture4']
Q1 = df_clean[moisture_cols].quantile(0.25)
Q3 = df_clean[moisture_cols].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df_clean[moisture_cols] < lower_bound) | (df_clean[moisture_cols] > upper_bound)).any(axis=1)
print(f"\nJumlah outliers: {outliers.sum()} dari {len(df_clean)}")
df_clean = df_clean[~outliers].copy()
print(f"Shape setelah hapus outliers: {df_clean.shape}")

time_cols = ['year', 'month', 'day', 'hour', 'minute', 'second']
df_clean['datetime'] = pd.to_datetime(df_clean[time_cols])
df_clean.set_index('datetime', inplace=True)
df_clean.drop(columns=time_cols, inplace=True)
print("Kolom datetime berhasil dibuat sebagai index.")

if 'irrgation' in df_clean.columns:
    df_clean['irrigation'] = df_clean['irrgation'].map({'false': 0, 'true': 1})
    df_clean.drop(columns=['irrgation'], inplace=True)
    print("Kolom irigasi dikonversi ke numerik.")

print("\nData setelah cleaning:")
print(df_clean.head())

moisture_cols = [col for col in df_clean.columns if 'moisture' in col]

print("\n=== Statistik Deskriptif ===")
print(df_clean[moisture_cols].describe())

plt.figure(figsize=(8,6))
corr = df_clean[moisture_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Korelasi Antar Sensor Kelembaban')
plt.tight_layout()
plt.show()

plt.figure(figsize=(14,6))
for col in moisture_cols:
    plt.plot(df_clean.index, df_clean[col], label=col, linewidth=0.8)
plt.xlabel('Waktu')
plt.ylabel('Nilai Kelembaban')
plt.title('Tren Kelembaban dari Waktu ke Waktu')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

df_clean[moisture_cols].hist(figsize=(15, 10), bins=30)
plt.suptitle('Histogram Nilai Kelembaban')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(data=df_clean[moisture_cols])
plt.title('Boxplot Sensor Kelembaban')
plt.ylabel('Nilai Kelembaban')
plt.grid(True)
plt.tight_layout()
plt.show()

output_file = 'soil_moisture_1_cleaned.csv'
df_clean.to_csv(output_file)
print(f"\nData bersih telah disimpan sebagai: {output_file}")
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Konfigurasi halaman
st.set_page_config(page_title="Dashboard Soil Moisture", layout="wide")
st.title("🌱 Dashboard Analisis Kelembaban Tanah (Soil Moisture)")

# ==============================
# 1. Baca data
# ==============================
@st.cache_data
def load_data():
    df = pd.read_csv('soil_moisture_1_cleaned.csv', index_col=0, parse_dates=True)
    return df

df = load_data()
moisture_cols = [col for col in df.columns if 'moisture' in col]

st.sidebar.header("Pengaturan")
st.sidebar.write(f"Data shape: {df.shape}")
st.sidebar.write(f"Rentang waktu: {df.index.min()} s/d {df.index.max()}")

# ==============================
# 2. Statistik Deskriptif
# ==============================
st.header("📊 Statistik Deskriptif")
st.dataframe(df[moisture_cols].describe())

# ==============================
# 3. Korelasi Antar Sensor
# ==============================
st.header("🔗 Korelasi Antar Sensor")
fig, ax = plt.subplots(figsize=(8, 6))
corr = df[moisture_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, ax=ax)
ax.set_title('Matriks Korelasi')
st.pyplot(fig)

# ==============================
# 4. Tren Time Series
# ==============================
st.header("📈 Tren Kelembaban dari Waktu ke Waktu")

# Pilih sensor yang ingin ditampilkan
selected_sensors = st.multiselect(
    "Pilih sensor yang akan ditampilkan",
    options=moisture_cols,
    default=moisture_cols
)

if selected_sensors:
    fig, ax = plt.subplots(figsize=(12, 6))
    for col in selected_sensors:
        ax.plot(df.index, df[col], label=col, linewidth=1)
    ax.set_xlabel('Waktu')
    ax.set_ylabel('Nilai Kelembaban')
    ax.set_title('Time Series')
    ax.legend()
    ax.grid(True)
    st.pyplot(fig)
else:
    st.warning("Pilih setidaknya satu sensor.")

# Opsi untuk melihat rentang waktu tertentu
st.subheader("Filter Rentang Waktu")
min_date = df.index.min().date()
max_date = df.index.max().date()
start_date = st.date_input("Mulai", min_date)
end_date = st.date_input("Selesai", max_date)

if start_date <= end_date:
    mask = (df.index.date >= start_date) & (df.index.date <= end_date)
    df_filtered = df.loc[mask]
    if not df_filtered.empty:
        fig2, ax2 = plt.subplots(figsize=(12, 4))
        for col in selected_sensors:
            ax2.plot(df_filtered.index, df_filtered[col], label=col, linewidth=1)
        ax2.set_xlabel('Waktu')
        ax2.set_ylabel('Kelembaban')
        ax2.set_title(f'Time Series ({start_date} s/d {end_date})')
        ax2.legend()
        ax2.grid(True)
        st.pyplot(fig2)
    else:
        st.warning("Tidak ada data pada rentang tersebut.")
else:
    st.error("Tanggal akhir harus setelah tanggal mulai.")

# ==============================
# 5. Distribusi Data
# ==============================
st.header("📊 Distribusi Nilai Kelembaban")

# Histogram
st.subheader("Histogram")
fig_hist, ax_hist = plt.subplots(figsize=(12, 8))
df[moisture_cols].hist(bins=30, ax=ax_hist, layout=(3, 2))
ax_hist.set_suptitle('Histogram')
plt.tight_layout()
st.pyplot(fig_hist)

# Boxplot
st.subheader("Boxplot")
fig_box, ax_box = plt.subplots(figsize=(10, 6))
sns.boxplot(data=df[moisture_cols], ax=ax_box)
ax_box.set_title('Boxplot Sensor')
ax_box.set_ylabel('Nilai Kelembaban')
ax_box.grid(True)
st.pyplot(fig_box)

# ==============================
# 6. Data Mentah (Opsional)
# ==============================
if st.checkbox("Tampilkan data mentah"):
    st.subheader("Data Cleaned")
    st.dataframe(df)

# ==============================
# 7. Footer
# ==============================
st.markdown("---")
st.markdown("Dashboard dibuat dengan ❤️ menggunakan Streamlit")
```
